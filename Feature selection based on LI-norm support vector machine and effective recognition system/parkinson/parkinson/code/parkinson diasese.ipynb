```python
scaler=MinMaxScaler()
df.iloc[:,1:-1]=scaler.fit_transform(df.iloc[:,1:-1])
```

```python
X=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

```python
import numpy
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

test = SelectKBest(score_func=chi2, k='all')
fit = test.fit(np.abs(X), y)
numpy.set_printoptions(precision=3)
print(fit.scores_)
```

```python
import itertools
feature_map = [(i, v) for i, v in itertools.zip_longest(list(df.columns[range(1,23)]),fit.scores_)]
```

```python
d=dict(feature_map)
```

```python
weighted_columns=sorted(d)
```

```python
X=df[weighted_columns]
```

```python
y.value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler 
ros = RandomOverSampler(random_state=42)
X_res, y_res = ros.fit_sample(X, y)
```

```python
from sklearn.metrics import accuracy_score
def fun(c_val,g,k):
    C=c_val
    kernel=k
    gm=g
    svm=SVC(kernel=k,C=C,gamma=gm)
    y_pred = cross_val_predict(svm, X_res, y_res , cv=10)

    accuracy =accuracy_score(y_res,y_pred)
    f1 = cross_val_score(svm, X_res, y_res, cv=10,scoring='f1')
    precission = cross_val_score(svm,X_res, y_res, cv=10,scoring='precision')
    recall=cross_val_score(svm, X_res, y_res, cv=10,scoring='recall')
    roc_auc = cross_val_score(svm,X_res, y_res, cv=10,scoring='roc_auc')
    return accuracy,f1,precission,recall,roc_auc
```

```python
accuracy,f1,precission,recall,roc_auc=fun(1,0.015,'rbf')
l1=['rbf',1,0.015,accuracy.mean(),f1.mean(),precission.mean(),recall.mean()]
```
